# Decision Tree Regression

Imagine a dataset where there are two independent variables x1 and x2 and they can be represented in scatter plot. The dependent variable y is in the third dimension. 

The scatter plot of x1 and x2 can be split into different sections of data (also known as leaves). The splitting of the data is dependent on the information entropy (research).

The splits can be drawn via a decision tree. i.e. is x1 > 20 yes or no...if no then is x2 < 170..etc

To predict a value of y for a new value, you check which leaf the new value falls into and then calculate the average y value from that leaf.

Will use the position salaries data again.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
X.shape, y.shape
```

```python
# Fit Decision tree regressor to the dataset
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
# predicting salary of level 6.5
y_pred = regressor.predict(6.5)
y_pred
```

```python
# visualise data
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Small sample size - only predicting 10 values
Non-continuous regression model
Need to look at results in smaller intervals/higher resolution

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

This is the real decision tree results. Not very interesting with one independent variables

Much better in more dimensions.
